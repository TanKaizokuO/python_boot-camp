```python
def my_dec(func):
    def wrapper():
        print("Before the function call")
        func()
        print("After the function call")
    return wrapper

@my_dec
def say_hello():
    print("Hello, World!")

say_hello()
```

### LOGGING USING DECORATORS

```python
from functools import wraps  # helps preserve original function metadata

def log_activity(func):      # decorator function that takes another function
    @wraps(func)             # keeps original name/docstring of the function
    def wrapper(*args, **kwargs):  # wrapper can accept any arguments
        print(f"🚀 Calling: {func.__name__}")  # log before function runs
        
        result = func(*args, **kwargs)  # call the original function
        
        print(f"✅ Finished: {func.__name__}")  # log after function runs
        
        return result  # return the original function result
    
    return wrapper  # return the wrapped function


@log_activity   # decorator applied to brew_chai
def brew_chai(type, milk="no"):
    print(f"Brewing {type} chai and milk status {milk}")


brew_chai("Masala")  # calling the decorated function
```

### AUTH USING DECORATORS

```python
from functools import wraps  # used to preserve original function metadata

def require_admin(func):  # decorator that checks if the user is admin
    @wraps(func)
    def wrapper(user_role):  # wrapper replaces the original function
        if user_role != "admin":  # check if role is not admin
            print("Access denied: Admins only")  # block access
            return None
        else:
            return func(user_role)  # call the original function if admin
    return wrapper  # return the wrapped function


@require_admin  # apply the decorator
def acess_tea_inventory(role):
    print("Access granted to tea inventory")


acess_tea_inventory("user")   # user is not admin
acess_tea_inventory("admin")  # admin is allowed
```
