#### Infinite Generator

```python
def infinite_numbers():
    n = 1
    while True:
        yield n
        n += 1

gen = infinite_numbers()

for _ in range(5):  # '_' means we don't care about the loop variable
    print("Hello")
    print(next(gen), end=' ')
```

#### Yield to take input
- anime_recommender() is a generator function because it uses yield.
- anime = yield pauses the function and waits for input.
- next(system) starts the generator until the first yield.
- system.send("Naruto") sends "Naruto" into the generator, stored in anime.
- The loop prints the recommendation and pauses again waiting for the next send().

```python
def anime_recommender():
    print("Welcome! Which anime would you like a recommendation for?")
    
    anime = yield   # wait for first request
    
    while True:
        print(f"Recommending anime similar to: {anime}")
        anime = yield   # wait for next request


# create generator
system = anime_recommender()

next(system)  # start the generator

system.send("Naruto")
system.send("Attack on Titan")
system.send("Death Note")
```

```python
def anime_list():
    yield "Naruto"
    yield "One Piece"
    yield "Bleach"

def anime_names():
    yield from anime_list()
    
for anime in anime_names():
    print(anime)
```

```python
def anime_stream():
    try:
        while True:  # run forever until the generator is closed
            anime = yield "Waiting for anime request..."  
            # pause here and return the message to the caller
            # when .send(value) is called, that value is stored in 'anime'

            print(f"Streaming episode of: {anime}")  
            # prints the anime name received from .send()

    except GeneratorExit:  
        # this exception happens when generator.close() is called
        print("Streaming service stopped.")  # cleanup message


stream = anime_stream()  
# create the generator object (generator not started yet)

print(next(stream))          
# start the generator until the first yield
# output: "Waiting for anime request..."

stream.send("Attack on Titan")  
# sends value into the generator
# 'anime' becomes "Attack on Titan"
# prints: Streaming episode of: Attack on Titan

stream.send("Jujutsu Kaisen")  
# sends another anime name
# prints: Streaming episode of: Jujutsu Kaisen

stream.close()               
# stops the generator
# raises GeneratorExit inside the generator
# prints: Streaming service stopped.
```
