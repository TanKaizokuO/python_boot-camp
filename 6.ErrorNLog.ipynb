```python
try:
    x = 0
    result = 10 / x
    print(result)

except ValueError as e:
    print("Invalid input. Please enter a number.")
    print("Error:", e)

except ZeroDivisionError as e:
    print("Cannot divide by zero.")
    print("Error:", e)
    
finally:
    print("Execution finished.")
```

```python
def summon_power(power):
    try:
        print(f"Charging {power}...")
        
        if power == "Magic":
            raise ValueError("Magic is forbidden in this realm!")

    except ValueError as e:
        print("Sensei says:", e)

    else:
        print(f"{power} unleashed!")

    finally:
        print("Next warrior, step forward.\n")


summon_power("Haki")
summon_power("Magic")
```

```python
def process_order(item, quantity):
    try:
        #{"masala": 20}[item] : is item in {...} ie the dictionary
        price = {"masala": 20}[item]
        cost = price * quantity
        print(f"total cost is {cost}")

    except KeyError:
        print("Sorry that chai is not on menu")

    except TypeError:
        print("Quantity must be in number")


process_order("ginger", 2)
process_order("masala", "two")
```

```python
class OutOfManaError(Exception):
    pass


def cast_spell(mana, focus):
    if mana == 0 or focus == 0:
        raise OutOfManaError("Not enough mana or focus to cast the spell!")

    print("✨ The spell has been successfully cast!")


cast_spell(0, 1)
```

```python
class UnknownJutsuError(Exception):
    pass


def mission_cost(jutsu, uses):
    scroll = {"fireball": 20, "shadow_clone": 40}

    try:
        if jutsu not in scroll:
            raise UnknownJutsuError("That jutsu is forbidden or unknown")

        if not isinstance(uses, int):
            raise TypeError("Number of uses must be an integer")

        chakra_cost = scroll[jutsu] * uses
        print(f"Chakra required for {uses} uses of {jutsu}: {chakra_cost}")

    except Exception as e:
        print("Sensei says:", e)

    finally:
        print("Mission evaluation complete.\n")


mission_cost("fireball", 2)
mission_cost("rasengan", 1)
mission_cost("shadow_clone", "two")
```
