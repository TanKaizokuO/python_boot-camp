## Object Oriented Programming

### **\_\_init\_\_ in Python**

\_\_init\_\_ method in Python is a constructor. It runs automatically when a new object of a class is created. Its main purpose is to initialize the object’s attributes and set up its initial state. When an object is created, memory is allocated for it, and \_\_init\_\_ helps organize that memory by assigning values to attributes.

#### Define a class

```python
class Animal:
    kingdom = "Animalia"   # class attribute
    domesticated = True     # class attribute

    # Constructor
    def __init__(self, name, species):
        self.name = name
        self.species = species

    # Method
    def speak(self):
        print(self.name + " makes a sound.")
 
    def can_domesticate(self):
        print(self.name + " can be domesticated: " + str(self.domesticated))
```

#### Create objects (instances of the class)

```python
dog = Animal("Shiro", "Dog")
cat = Animal("Whiskers", "Cat")
```

#### Access attributes

```python
print(dog.name)
print(cat.species)
```

#### Call methods

```python
dog.speak()
cat.speak()
```

#### Access class attribute

```python
print("Dog kingdom:", dog.kingdom)
print("Cat kingdom:", cat.kingdom)
```

#### Attribute shadowing

```python
dog.kingdom = "Modified Kingdom"   # instance attribute shadows class attribute

print("\nAfter shadowing:")

print("Dog kingdom:", dog.kingdom)        # instance attribute
print("Cat kingdom:", cat.kingdom)        # still class attribute
print("Class kingdom:", Animal.kingdom)   # class attribute unchanged


del dog.kingdom
print("Dog kingdom:", dog.kingdom)   # again uses class attribute
```

```python
dog.hindi = "Kutta" # Adding a new instance attribute
print("Dog's Hindi name:", dog.hindi)


del dog.hindi
try :
    print("Dog's Hindi name after deletion:", dog.hindi)  # This will raise an AttributeError since hindi is deleted
except AttributeError:
    print("Dog's Hindi name has been deleted and is no longer accessible.")
```

### Self keyword
self is a fundamental concept when working with object-oriented programming (OOP). It represents the instance of the class being used. Whenever we create an object from a class, self refers to the current object instance. It is essential for accessing attributes and methods within the class.

```python
dog.can_domesticate()

tiger = Animal("Sheru", "Tiger")
tiger.domesticated = False  # Override class attribute for this instance
tiger.can_domesticate()
```

**We can call methods using class name and by passing instance**

```python
Animal.can_domesticate(dog)  # Calling method using class name and passing instance
```
