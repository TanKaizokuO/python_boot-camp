## Object Oriented Programming

#### Define a class

```python
class Animal:
    kingdom = "Animalia"   # class attribute

    # Constructor
    def __init__(self, name, species):
        self.name = name
        self.species = species

    # Method
    def speak(self):
        print(self.name + " makes a sound.")
```

#### Create objects (instances of the class)

```python
dog = Animal("Shiro", "Dog")
cat = Animal("Whiskers", "Cat")
```

#### Access attributes

```python
print(dog.name)
print(cat.species)
```

#### Call methods

```python
dog.speak()
cat.speak()
```

#### Access class attribute

```python
print("Dog kingdom:", dog.kingdom)
print("Cat kingdom:", cat.kingdom)
```

#### Attribute shadowing

```python
dog.kingdom = "Modified Kingdom"   # instance attribute shadows class attribute

print("\nAfter shadowing:")

print("Dog kingdom:", dog.kingdom)        # instance attribute
print("Cat kingdom:", cat.kingdom)        # still class attribute
print("Class kingdom:", Animal.kingdom)   # class attribute unchanged


del dog.kingdom
print("Dog kingdom:", dog.kingdom)   # again uses class attribute
```

```python
dog.hindi = "Kutta" # Adding a new instance attribute
print("Dog's Hindi name:", dog.hindi)


del dog.hindi
try :
    print("Dog's Hindi name after deletion:", dog.hindi)  # This will raise an AttributeError since hindi is deleted
except AttributeError:
    print("Dog's Hindi name has been deleted and is no longer accessible.")
```
