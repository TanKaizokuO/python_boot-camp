| Aspect                       | Inheritance                                                                     | Composition                                                         |
| ---------------------------- | ------------------------------------------------------------------------------- | ------------------------------------------------------------------- |
| **Definition**               | A class **derives from another class** and inherits its attributes and methods. | A class **contains objects of another class** as components.        |
| **Purpose**                  | Used to **reuse and extend behavior** of an existing class.                     | Used to **build complex objects from simpler objects**.             |
| **Dependency**               | Child class is **tightly coupled** to the parent class.                         | Classes are **loosely coupled**; components can be replaced easily. |
| **Code Reuse Mechanism**     | Reuse through **class hierarchy**.                                              | Reuse through **object composition**.                               |
| **Flexibility**              | Less flexible; changing the parent class can affect all child classes.          | More flexible; components can be swapped or modified independently. |
| **Implementation in Python** | `class Child(Parent):`                                                          | `self.component = Component()`                                      |
| **Example Concept**          | `Ninja is-a AnimeCharacter`                                                     | `AnimeCharacter has-a Weapon`                                       |                                              |
| **Typical Use Case**         | When classes share **common behavior and identity**.                            | When an object **uses another object to perform tasks**.            |


````python
# Parent class
class AnimeCharacter:
    weapon_cls = Weapon   # class attribute used for dependency injection

    def __init__(self, name, anime, weapon_name, power):
        self.name = name
        self.anime = anime
        self.weapon = self.weapon_cls(weapon_name, power)

    def introduce(self):
        print(self.name, "is from", self.anime)
        self.weapon.show_weapon()
````

````python
# Component class
class Weapon:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def show_weapon(self):
        print("Weapon:", self.name, "| Power:", self.power, end=" \n\n")
````

````python
# Specialized weapon
class LegendaryWeapon(Weapon):
    def show_weapon(self):
        print("Legendary Weapon:", self.name, "| Power:", self.power, end=" \n\n")
````

````python
# Child class
class Ninja(AnimeCharacter):
    weapon_cls = Weapon   # uses normal weapon

    def __init__(self, name, anime, jutsu, weapon_name, power):
        super().__init__(name, anime, weapon_name, power)
        self.jutsu = jutsu

    def use_jutsu(self):
        print(self.name, "uses", self.jutsu)
````

````python
# Another child class
class Pirate(AnimeCharacter):
    weapon_cls = LegendaryWeapon   # inject different dependency

    def __init__(self, name, anime, crew, weapon_name, power):
        super().__init__(name, anime, weapon_name, power)
        self.crew = crew

    def show_crew(self):
        print(self.name, "belongs to", self.crew, end=" \n\n")
````

````python
# Create character objects
naruto = Ninja("Naruto", "Naruto", "Shadow Clone Jutsu", "Kunai", 70)
luffy = Pirate("Luffy", "One Piece", "Straw Hat Pirates", "Gomu Punch", 95)

# Using inherited method
naruto.introduce()
luffy.introduce()

# Child-specific behavior
naruto.use_jutsu()
luffy.show_crew()
````

## **Accessing Base Class**

| Method               | One-line Explanation                                                                                                      |
| -------------------- | ------------------------------------------------------------------------------------------------------------------------- |
| **Code Duplication** | The child class manually re-implements the logic of the base class instead of calling it.                                 |
| **Explicit Call**    | The child class directly calls the base class method using `BaseClass.method(self, ...)`.                                 |
| **super()**          | The recommended Python way to access base class methods using `super()` which respects the method resolution order (MRO). |


````python
class Chai:
    def __init__(self, type_, strength):
        self.type = type_
        self.strength = strength
````

### 1. Code Duplication

````python
class LemonChai(Chai):
    def __init__(self, type_, strength, spice_level):
        self.type = type_
        self.strength = strength
        self.spice_level = spice_level
````

### 2. Explicit Call

````python
class GreenChai(Chai):
    def __init__(self, type_, strength, spice_level):
        Chai.__init__(self, type_, strength)
        self.spice_level = spice_level
````

### 3. Super() Keyword {RECOMMENDED}

````python

class GingerChai(Chai):
    def __init__(self, type_, strength, spice_level):
        super().__init__(type_, strength)
        self.spice_level = spice_level
````

### Method Resolution Order (MRO) — Summary

| Aspect                    | Description                                                                                    |
| ------------------------- | ---------------------------------------------------------------------------------------------- |
| **Definition**            | The order in which Python searches classes to resolve a method or attribute.                   |
| **Purpose**               | Ensures Python knows **which method to execute** when multiple classes define the same method. |
| **Where Used**            | Primarily important in **inheritance and multiple inheritance** scenarios.                     |
| **Search Order**          | Python looks for methods in: **instance → current class → parent classes → object class**.     |
| **Algorithm Used**        | Python uses the **C3 Linearization algorithm** to determine a consistent method lookup order.  |
| **Accessing MRO**         | Can be viewed using `ClassName.mro()` or `ClassName.__mro__`.                                  |
| **Relation to `super()`** | `super()` follows the **MRO chain** to call the next method in the resolution order.           |


````python
class A:
    def show(self):
        print("A")

class B(A):
    pass

class C(B):
    pass

print(C.mro())
````

Meaning Python searches methods in this order:

```
C → B → A → object
```

### Key Idea

> **MRO defines the path Python follows to find methods in inheritance hierarchies.**


````python
class A:
    label = "Class A"

class B(A):
    label = "Class B"

class C(A):
    label = "Class C"

class D(B, C):#first B, then C
    pass

class E(C, B):#first C, then B
    pass

print("D.__mro__:", D.__mro__)
print("E.__mro__:", E.__mro__)
````
