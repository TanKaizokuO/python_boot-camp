## @staticmethod

A static method in Python is a method defined inside a class that does not depend on any instance or class data. It is used when a function logically belongs to a class but does not need access to self or cls. Static methods help organize related utility functions inside a class without creating objects.

```python
class Calc:
    @staticmethod
    def add(a, b):
        return a + b

res = Calc.add(2, 3)
print(res)
```

## classmethod() 

The classmethod() is an inbuilt function in Python, which returns a class method for a given function. This means that classmethod() is a built-in Python function that transforms a regular method into a class method. When a method is defined using the @classmethod decorator (which internally calls classmethod()), the method is bound to the class and not to an instance of the class. As a result, the method receives the class (cls) as its first argument, rather than an instance (self).

| Aspect                       | `@classmethod`                                                          | `@staticmethod`                                                               |
| ---------------------------- | ----------------------------------------------------------------------- | ----------------------------------------------------------------------------- |
| **Definition**               | A method that receives the **class (`cls`) as the first argument**.     | A method that **does not receive class or instance automatically**.           |
| **Access to Class State**    | Can **access and modify class attributes**.                             | Cannot directly access class or instance attributes unless passed explicitly. |
| **Access to Instance State** | Cannot directly access instance attributes.                             | Cannot access instance attributes.                                            |
| **Purpose**                  | Used for **factory methods or operations related to the class itself**. | Used for **utility/helper functions logically grouped inside the class**.     |


```python
class Chai:

    shop_name = "Chai Corner"

    def __init__(self, type_, strength):
        self.type = type_
        self.strength = strength

    @classmethod
    def from_dict(cls, info):
        return cls(info["type"], info["strength"])

    @classmethod
    def show_shop(cls):
        print("Shop:", cls.shop_name)

    @staticmethod
    def greet():
        print("Welcome to the chai shop!")

greenTea = Chai.from_dict({"type": "Green", "strength": "Medium"})
print("Type:", greenTea.type)
print("Strength:", greenTea.strength)

Chai.show_shop()
Chai.greet()
```

| Concept         | Explanation                                                                                        |
| --------------- | -------------------------------------------------------------------------------------------------- |
| **`@property`** | A decorator that allows a method to be accessed like an **attribute** instead of a function call.  |
| **Getter**      | A method that **retrieves the value** of an attribute in a controlled way.                         |
| **Setter**      | A method that **updates the value** of an attribute while applying validation or additional logic. |


```python
class TeaLeaf:
    def __init__(self, age):
        self.age = age   # use setter for validation

    @property
    def age(self):
        """Return processed tea leaf age."""
        return self._age + 2

    @age.setter
    def age(self, value):
        """Validate and store raw tea leaf age."""
        if 1 <= value <= 5:
            self._age = value
        else:
            raise ValueError("Tea leaf age must be between 1 and 5 years")


# Create object
leaf = TeaLeaf(2)

# Access property
print("Processed age:", leaf.age)

# Update value
leaf.age = 4
print("Updated processed age:", leaf.age)

# Invalid update
try:
    leaf.age = 6
except ValueError as e:
    print("Error:", e)
```
