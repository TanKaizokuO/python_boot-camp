```python
anime_episodes = [
    ["Naruto", 220],
    ["Attack on Titan", 94],
    ["Death Note", 37],
    ["Fullmetal Alchemist: Brotherhood", 64],
    ["One Piece", 1100],
    ["Demon Slayer", 55],
    ["My Hero Academia", 138],
    ["Jujutsu Kaisen", 47],
    ["Dragon Ball Z", 291],
    ["Tokyo Ghoul", 48],
    ["Steins;Gate", 24],
    ["Hunter x Hunter", 148],
    ["Sword Art Online", 96],
    ["Code Geass", 50],
    ["Bleach", 366]
]
```

```python
# Dictionary mapping genre -> anime names
anime_by_genre = {
    "Action": [
        "Naruto",
        "Attack on Titan",
        "Demon Slayer",
        "Jujutsu Kaisen",
        "Dragon Ball Z",
        "Bleach",
    ],
    "Adventure": [
        "One Piece",
        "Hunter x Hunter",
        "Sword Art Online",
        "Fullmetal Alchemist: Brotherhood",
    ],
    "Psychological": ["Death Note", "Steins;Gate", "Tokyo Ghoul", "Code Geass"],
    "Superhero": ["My Hero Academia"],
    "Dark Fantasy": ["Attack on Titan", "Tokyo Ghoul", "Demon Slayer"],
}
```

```python
anime_watch_list = [anime_info[0] for anime_info in anime_episodes if anime_info[1] < 50]
print("Anime to watch (less than 50 episodes):")
for anime in anime_watch_list:
    print(anime)
```

```python
shonen_watchlist = [anime for genre in anime_by_genre for anime in anime_by_genre[genre] if genre == "Action" ]
print("\nShonen anime to watch:")
for anime in shonen_watchlist:
    print(anime)
```
