### Learning Enumerate function

enumerate() function in Python is used to loop over an iterable and get both the index and the element at the same time. It returns an enumerate object that produces pairs in the form (index, element). This removes the need to manually maintain a counter variable during iteration.

```python
fruits = ["apple", "banana", "cherry"]
print(list(enumerate(fruits,start=1)))
```

```python
tasks = ["apple", "banana", "cherry"]
tasks = list(enumerate(tasks, start=1))
print([f"{x[0]}. {x[1]}" for x in tasks])
```

### Zip Function

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together etc.

```python
manga = ["One Piece", "Naruto", "Bleach", "Dragon Ball", "Attack on Titan"]
mangaka = ["Eiichiro Oda", "Masashi Kishimoto", "Tite Kubo", "Akira Toriyama", "Hajime Isayama"]
manga_mangaka = list(zip(manga, mangaka))
print(manga_mangaka)
```

### Walrus operator

The Walrus Operator (:=), introduced in Python 3.8, allows you to assign a value to a variable as part of an expression. It helps avoid redundant code when a value needs to be both used and tested in the same expression — especially in loops or conditional statements.

```python
number = 14
if(number2 := number % 3):
    print(f"{number} is not divisible by 3, the remainder is {number2}")
```

```python
d = [
    {"userId": 1, "name": "rahul", "completed": False},
    {"userId": 1, "name": "rohit", "completed": False},
    {"userId": 1, "name": "ram", "completed": False},
    {"userId": 1, "name": "ravan", "completed": True}
]

print("With Python 3.8 Walrus Operator:")
for entry in d:
    if name := entry.get("name"):
        print(name)
print("\n")
print("Without Walrus operator:")
for entry in d:
    name = entry.get("name")
    if name:
        print(name)
```

```python
users = [
    {"id": 1, "total": 100, "coupon": "P20"},
    {"id": 2, "total": 150, "coupon": "F10"},
    {"id": 3, "total": 80, "coupon": "P50"},
    {"id": 4, "total": 200, "coupon": "F30"},
]

discounts = {
    "P20": (0.2, 0),
    "F10": (0.5, 0),
    "P50": (0, 10),
}

for user in users:
    percent, fixed = discounts.get(user["coupon"], (0, 0))
    # print("Percent:", percent, "Fixed:", fixed)
    # Percent: 0.2 Fixed: 0
    # Percent: 0.5 Fixed: 0
    # Percent: 0 Fixed: 10
    # Percent: 0 Fixed: 0
    discount = user["total"] * percent + fixed
    print(f"{user["id"]} paid {user["total"]} and got discount for next visit of rupees {discount}")
```

### Scopes and Named Space

```python
def order_food(quantity=1):
    order = "chicken biriyani"
    def print_order():
        order = "Mutton Biriyani" #local scope of print_order function
        print(f"Order: {order}, Quantity: {quantity}")
    print_order()
    print(f"Order: {order}, Quantity: {quantity}")  

order="Hyderabadi biriyani" #global scope
order_food()
print(f"Order: {order}")
```

```python
def RAG(chunk1, chunk2):
    similarity = 1
    print(f"Similarity before calculate_similarity: {similarity}")
    def calculate_similarity():
        # # similarity = 0.5 THIS GAVE AN EEROR "name 'similarity' is used prior to nonlocal declaration"
        # print(f"Similarity inside calculate_similarity: {similarity}")
        nonlocal similarity
        similarity = 0.8
        print(f"Similarity after calculate_similarity: {similarity}")
    calculate_similarity()
    print(f"Similarity after calculate_similarity in RAG: {similarity}")
RAG("chunk1", "chunk2")
```

```python
value = "global"
g = "initial global"

def outer():
    value = "enclosing"

    def inner():
        nonlocal value
        global g
        value = "nonlocal updated"
        g = "global updated"

    inner()
    print("value inside outer :", value)
    print()


print("global g before calling outer:", g)
print("value before calling outer:", value)
print()
outer()
print("value after calling outer:", value)
print("g after calling outer:", g)
```

**Remember in python non primitive data types are paased as call by reference**

```python
def increase_length(a): #Parameter a is a reference to the original list passed as an argument
    a.append("new element")
    # print(len(a))

def update_array(arr, n, i):
    arr[i-1] = n

my_list = [1, 2, 3]
print("Before calling increase_length:", my_list)
increase_length(my_list)   #Argument my_list is passed by reference, so the original list is modified
print("After calling increase_length:", my_list)
print("\n")
my_array = [0, 0, 0, 0, 0]
print("Before calling update_array:", my_array)
update_array(my_array, 5, 3)
print("After calling update_array:", my_array)
```

**Value passing using Keyword**

```python
def print_records(name, age, city):
    print(f"Name: {name}, Age: {age}, City: {city}")

print_records("Alice", 30, "New York") #positional arguments
print_records(name="Bob", age=25, city="Los Angeles") #keyword arguments, order of arguments can be changed when using keyword arguments
```

```python
def special_chain(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

special_chain(1, 2, 3, name="Alice", age=30)

# **kwargs — Variable Keyword Arguments
# Collects extra named arguments into a dictionary
# Lets you pass any number of key-value pairs

#*args — Variable Positional Arguments
# Collects extra positional arguments into a tuple
# Lets you pass any number of unnamed inputs
```

IF YOU WANT TO PASS A EMPTY LIST TO A FUNCTION USE **def func(arr=None):** instead of **arr=[]** 

### Returning multiple values from a function 

```python
def what_anime():
    return "One piece", "Naruto"

anime = what_anime() # anime is a tuple containing the returned values from the function
anime1, anime2 = what_anime() # anime1 and anime2 are assigned the values returned by the function, unpacking the tuple into separate variables
print("Anime:", anime)
print("Anime 1:", anime1)
print("Anime 2:", anime2)
```
