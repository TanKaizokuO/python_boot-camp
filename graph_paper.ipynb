```python
import matplotlib.pyplot as plt
import numpy as np

# Methods
methods = ["Plain FL", "BatchCrypt", "Proposed HE (CKKS-Light)"]

# Data (approximate values from your figure)
accuracy = [76, 74, 82]
communication = [120, 480, 600]   # KB per round
encryption_time = [0.0, 0.25, 0.01]  # seconds

# Create vertical layout
fig, axes = plt.subplots(3, 1, figsize=(6, 10))

# --- Accuracy ---
axes[0].bar(methods, accuracy)
axes[0].set_title("Model Accuracy Comparison")
axes[0].set_ylabel("Accuracy (%)")
axes[0].set_ylim(0, 90)

# --- Communication Overhead ---
axes[1].bar(methods, communication)
axes[1].set_title("Communication Overhead per Round")
axes[1].set_ylabel("Communication (KB)")

# --- Encryption Time ---
axes[2].bar(methods, encryption_time)
axes[2].set_title("Encryption + Decryption Overhead")
axes[2].set_ylabel("Time (seconds)")

# Rotate labels for readability
for ax in axes:
    ax.tick_params(axis='x', rotation=20)

# Compact spacing
plt.tight_layout()

# Save high-resolution figure
plt.savefig("fl_results_vertical.png", dpi=300, bbox_inches="tight")

plt.show()
```
